```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import csv

#import requests
#r = requests.get('https://drive.google.com/file/d/1l6duNna670RrZcvrZl1MFHLm5b1uORLi')
#data = r.content
#print (data)

df = pd.read_csv (r'/content/weather22.csv')
#print (df)

x = []
y = []
  
with open('/content/weather22.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[0])
        y.append(row[1])
  
plt.bar(x, y, color = 'r', width = 1, label = "average temp")
plt.xlabel('Temp')
plt.ylabel('Date')
plt.title('weather Averages for 2022')
plt.legend()
plt.show()

print("____________________________________________________________________________________________")

Names = []
Values = []
  
with open('/content/weather22.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        Names.append(row[0])
        Values.append(row[1])
  
plt.scatter(Names, Values, color = 'g',s = 1)
plt.xticks(rotation = 30)
plt.xlabel('date')
plt.ylabel('temps')
plt.title('Average Temps 2022', fontsize = 20)
  
plt.show()




#url = pd.read_csv (r'https://drive.google.com/file/d/1l6duNna670RrZcvrZl1MFHLm5b1uORLi/view?usp=sharing')
#print (df)

#google drive link
#https://drive.google.com/drive/folders/1-rID-nSaFAJ52bViAClazL-a70cXvKON
```
